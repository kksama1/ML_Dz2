```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.base import BaseEstimator, RegressorMixin, TransformerMixin
```

```python
df = pd.read_csv("regularization.csv")
df
```

```python
class LinearRegressionWithRegularization(BaseEstimator, RegressorMixin):
    def __init__(self, alpha=0.0):
        self.alpha = alpha
    
    def fit(self, X, y):
        # Добавление регуляризации (Ridge Regression)
        X_transpose = np.transpose(X)
        identity = np.identity(X.shape[1])
        identity[0, 0] = 0  # Не регуляризуем перехват
        regularization_matrix = self.alpha * identity
        self.coef_ = np.linalg.lstsq(np.dot(X_transpose, X) + regularization_matrix, np.dot(X_transpose, y), rcond=None)[0]
        self.intercept_ = 0  # Перехват установлен в 0
    
    def predict(self, X):
        return np.dot(X, self.coef_)
    
    def get_params(self, deep=True):
        return {"alpha": self.alpha}
```

```python
class StandardScalerCustom(BaseEstimator, TransformerMixin):
    def __init__(self, has_bias=False, apply_mean=True):
        self.has_bias = has_bias
        self.apply_mean = apply_mean
    
    def fit(self, X, y=None):
        if self.has_bias:
            X = X[:, 1:]
        self.mean_ = np.mean(X, axis=0) if self.apply_mean else np.zeros(X.shape[1])
        self.scale_ = np.std(X, axis=0, ddof=0)
        return self
    
    def transform(self, X):
        if self.has_bias:
            X = X[:, 1:]
        return (X - self.mean_) / self.scale_
    
    def fit_transform(self, X, y=None):
        return self.fit(X, y).transform(X)
```

```python
def run_holdout(model, X, y, train_size, random_state) -> dict:
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_size, random_state=random_state)
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    mse_train = mean_squared_error(y_train, y_pred_train)
    r2_train = r2_score(y_train, y_pred_train)
    mse_test = mean_squared_error(y_test, y_pred_test)
    r2_test = r2_score(y_test, y_pred_test)
    
    return {
        'mse_train': mse_train,
        'r2_train': r2_train,
        'mse_test': mse_test,
        'r2_test': r2_test
    }

def run_cross_val(model, X, y, n_splits, shuffle, random_state) -> dict:
    kf = KFold(n_splits=n_splits, shuffle=shuffle, random_state=random_state)
    mse_train_scores = []
    r2_train_scores = []
    mse_test_scores = []
    r2_test_scores = []
    
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        model.fit(X_train, y_train)
        y_pred_train = model.predict(X_train)
        y_pred_test = model.predict(X_test)
        
        mse_train_scores.append(mean_squared_error(y_train, y_pred_train))
        r2_train_scores.append(r2_score(y_train, y_pred_train))
        mse_test_scores.append(mean_squared_error(y_test, y_pred_test))
        r2_test_scores.append(r2_score(y_test, y_pred_test))
    
    return {
        'mse_train': np.mean(mse_train_scores),
        'r2_train': np.mean(r2_train_scores),
        'mse_test': np.mean(mse_test_scores),
        'r2_test': np.mean(r2_test_scores)
    }
```

```python
# Разделение на X и y
X = df.drop('Y', axis=1).values
y = df['Y'].values

# Параметры для моделей
alphas = [0, 0.01]
train_size = 0.75
n_splits = 4
shuffle = True
random_state = 0

for alpha in alphas:
    print(f'Обучение модели с alpha = {alpha}')
    
    # Создание конвейера
    pipeline = Pipeline([
        ('scaler', StandardScalerCustom(has_bias=True, apply_mean=True)),  # Стандартизация данных
        ('regressor', LinearRegressionWithRegularization(alpha=alpha))  # Линейная регрессия с регуляризацией
    ])
    
    # Оценка модели
    holdout_scores = run_holdout(pipeline, X, y, train_size, random_state)
    cross_val_scores = run_cross_val(pipeline, X, y, n_splits, shuffle, random_state)
    
    print(f'Holdout MSE и R^2 для alpha={alpha}:', holdout_scores)
    print(f'Кросс-валидация MSE и R^2 для alpha={alpha}:', cross_val_scores)
    
    # Обучение модели на всех данных для визуализации
    pipeline.fit(X, y)
    y_pred = pipeline.predict(X)
    
    # Визуализация предсказаний
    plt.figure(figsize=(12, 6))
    
    # График предсказаний
    plt.subplot(1, 2, 1)
    plt.scatter(y, y_pred, color='blue', alpha=0.5)
    plt.xlabel('Действительное значение (y)')
    plt.ylabel('Предсказание (ŷ)')
    plt.title(f'Предсказания для alpha={alpha}')
    
    # Добавление диагональной линии
    plt.plot([min(y), max(y)], [min(y), max(y)], color='red', linestyle='--')
    
    plt.subplot(1, 2, 2)
    plt.scatter(range(len(y)), y - y_pred, color='blue', alpha=0.5)
    plt.xlabel('Индекс')
    plt.ylabel('Ошибка (y - ŷ)')
    plt.title(f'Ошибки предсказания для alpha={alpha}')
    
    plt.tight_layout()
    plt.show()

    # Вывод параметров модели
    coefficients = pipeline.named_steps['regressor'].coef_
    intercept = pipeline.named_steps['regressor'].intercept_

    print(f'Коэффициенты модели для alpha={alpha}:', coefficients)
    print(f'Перехват (intercept) модели для alpha={alpha}:', intercept)
```
